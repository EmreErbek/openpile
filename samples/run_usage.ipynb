```python
from openpile.construct import Pile, Model
from openpile.analyses import simple_beam_analysis

import numpy as np
import matplotlib.pyplot as plt

p = Pile.create_tubular(name="", top_elevation=10, bottom_elevation=0, diameter=0.3, wt=0.01)
m = Model(name="", pile=p)
m.set_support(elevation=10, Ty=True)
m.set_support(elevation=0, Ty=True, Tx=True)
m.set_pointload(elevation=5, Py=1)

list_Px = np.linspace(1, 2_000, 500)

yhalf = []
xtop = []

for load in list_Px:
    m.set_pointload(elevation=10, Px=-load)
    results = simple_beam_analysis(m)
    yhalf.append(results.deflection.iloc[10, 1])

plt.plot(yhalf, list_Px)
```
